# First/last activity

```python
query = f"""
WITH enc as (
    select CONCAT('SiteSchema', '_', CAST(patid as varchar)) AS syn_pt_id
        , min(CAST(admit_date AS DATE)) as first_activity
        , max(CAST(admit_date AS DATE)) as last_activity
    FROM "{database_name}"."SiteSchema_pcornet_all_encounter"
    group by CONCAT('SiteSchema', '_', CAST(patid as varchar))
),
dx as (
    select CONCAT('SiteSchema', '_', CAST(patid as varchar)) AS syn_pt_id
        , min(CAST(admit_date AS DATE)) as first_activity
        , max(CAST(admit_date AS DATE)) as last_activity
    FROM "{database_name}"."SiteSchema_pcornet_all_diagnosis"
    group by CONCAT('SiteSchema', '_', CAST(patid as varchar))
),
proc as (
    select CONCAT('SiteSchema', '_', CAST(patid as varchar)) AS syn_pt_id
        , min(CAST(px_date AS DATE)) as first_activity
        , max(CAST(px_date AS DATE)) as last_activity
    FROM "{database_name}"."SiteSchema_pcornet_all_procedures"
    group by CONCAT('SiteSchema', '_', CAST(patid as varchar))
),
disp as (
    select CONCAT('SiteSchema', '_', CAST(patid as varchar)) AS syn_pt_id
        , min(CAST(DISPENSE_DATE AS DATE)) as first_activity
        , max(CAST(DISPENSE_DATE AS DATE)) as last_activity
    FROM "{database_name}"."SiteSchema_pcornet_all_dispensing"
    group by CONCAT('SiteSchema', '_', CAST(patid as varchar))
),
lab as (
    select CONCAT('SiteSchema', '_', CAST(patid as varchar)) AS syn_pt_id
        , min(CAST(RESULT_DATE AS DATE)) as first_activity
        , max(CAST(RESULT_DATE AS DATE)) as last_activity
    FROM "{database_name}"."SiteSchema_pcornet_all_LAB_RESULT_CM"
    group by CONCAT('SiteSchema', '_', CAST(patid as varchar))
),
presc as (
    select CONCAT('SiteSchema', '_', CAST(patid as varchar)) AS syn_pt_id
        , min(CAST(RX_ORDER_DATE AS DATE)) as first_activity
        , max(CAST(RX_ORDER_DATE AS DATE)) as last_activity
    FROM "{database_name}"."SiteSchema_pcornet_all_PRESCRIBING"
    group by CONCAT('SiteSchema', '_', CAST(patid as varchar))
),
MED_ADMIN as (
    select CONCAT('SiteSchema', '_', CAST(patid as varchar)) AS syn_pt_id
        , min(CAST(MEDADMIN_START_DATE AS DATE)) as first_activity
        , max(CAST(MEDADMIN_START_DATE AS DATE)) as last_activity
    FROM "{database_name}"."SiteSchema_pcornet_all_MED_ADMIN"
    group by CONCAT('SiteSchema', '_', CAST(patid as varchar))
),
OBS_GEN as (
    select CONCAT('SiteSchema', '_', CAST(patid as varchar)) AS syn_pt_id
        , min(CAST(OBSGEN_START_DATE AS DATE)) as first_activity
        , max(CAST(OBSGEN_START_DATE AS DATE)) as last_activity
    FROM "{database_name}"."SiteSchema_pcornet_all_OBS_GEN"
    group by CONCAT('SiteSchema', '_', CAST(patid as varchar))
),
all_ranges as (
select * from enc
union
select * from dx
union
select * from proc
union
select * from disp
union
select * from lab
union
select * from presc
union
select * from MED_ADMIN
union
select * from OBS_GEN
)

select syn_pt_id
    , min(first_activity) as first_activity
    , max(last_activity) as last_activity
from all_ranges
group by syn_pt_id
"""

df = rcb.extract_data_athena(
    query=query,
    site_names=site_names,
    data_name='first_last_activity',
    source_data_path=source_data_path,
    region=region,
    output_s3_bucket=output_s3_bucket,
    database_name=database_name,
    athena_client=athena_client,
    workgroup_name=workgroup_name
)

del query
del df
```
